```python
from alpaca_trade_api.rest import REST
from dotenv import load_dotenv
import os
import datetime
import pytz
import time

# 加载 .env 文件
load_dotenv()

API_KEY = os.getenv("APCA_API_KEY_ID")
SECRET_KEY = os.getenv("APCA_API_SECRET_KEY")
BASE_URL = "https://paper-api.alpaca.markets"

"""
APCA_API_KEY_ID=PK7NHVAMHZPYIXK0F43X
APCA_API_SECRET_KEY=iX47VfT39FzGo3gnnM8a7mxGyRr6aAC098gnPozu
"""
# 初始化 API
api = REST(API_KEY, SECRET_KEY, BASE_URL)
```

```python
# 设置美国东部时区
eastern = pytz.timezone("US/Eastern")

def is_trading_day(today):
    cal = api.get_calendar(start=today.isoformat(), end=today.isoformat())
    return len(cal) > 0

def buy_qqq_now():
    try:
        print("⏳ 准备提交 QQQ 买单...")
        api.submit_order(
            symbol='QQQ',
            qty=1,
            side='buy',
            type='market',
            time_in_force='gtc'
        )
        print("✅ 已提交 1 股 QQQ 市价买单")
    except Exception as e:
        print("❌ 下单失败：", e)
```

```python
while True:
    now = datetime.datetime.now(eastern)

    # 取当前时间的字符串格式
    time_str = now.strftime('%H:%M')
    
    # 判断是否是交易日
    if is_trading_day(now.date()):
        if time_str == "15:50":
            print(f"🕒 {now.strftime('%Y-%m-%d %H:%M:%S')} 是交易日，触发下单")
            buy_qqq_now()
            time.sleep(61)  # 防止重复下单
        else:
            print(f"等待中：当前时间 {time_str}，不是 15:50")
    else:
        print(f"❌ {now.date()} 不是交易日，跳过")

    time.sleep(60)  # 每分钟轮询一次
```
